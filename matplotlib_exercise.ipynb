## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
#csvの取得
data = pd.read_csv('data/hist_data.csv')
#グラフのタイトル
plt.title('aデータとbデータの温度に関する度数分布')
#x・y軸の軸名
plt.xlabel('温度(℃)'),plt.ylabel('度数')
#範囲の指定
plt.xlim(0,52),plt.ylim(0,28)
#データAのヒストグラム
plt.hist(data['aデータ'],bins = 30,label = 'aデータ',alpha = 0.5,color = 'orange')
#データbのヒストグラム
plt.hist(data['bデータ'],bins = 30 ,label ='bデータ',alpha = 0.3,color = 'cyan')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
data = pd.read_csv('data/scatter_data.csv')
#タイトル
plt.title('クラス0~2のlengthとwidthに関する散布図')
#x軸とy軸
plt.xlabel('length(cm)'),plt.ylabel('width(cm)')
#x軸とy軸の値の範囲
plt.xlim(4.1,8.1),plt.ylim(1.9,4.52)
#目盛線の表示
plt.grid(True)
#各クラスのデータ収集
class0 = data[data['class'] == 0]
class1 = data[data['class'] == 1]
class2 = data[data['class'] == 2]
plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='r')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='class1', color='b')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='class2', color='g')
#散布図の表示
plt.legend()
plt.grid(True)
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
data = pd.read_csv('data/line_data.csv')
#グラフのタイトル1つ目
plt.title('期間ごとの件数の推移')
plt.xlabel('期間'),plt.ylabel('件数')
#x軸とy軸の設定
plt.xticks(rotation = 30),plt.ylim(103,144)
#目盛の表示
plt.grid(True)
#データのプロット
plt.plot(data['期間'],data['件数'],marker = 'o')
plt.show()

#グラフ2つ目
plt.title('期間ごとの費用の推移')
plt.xlabel('期間'),plt.ylabel('費用(円)')
#x軸の設定
plt.xticks(rotation = 30)
#目盛の表示
plt.grid(True)
#棒グラフの描画
plt.bar(data['期間'],data['費用'],width=0.8)
plt.show()
```
